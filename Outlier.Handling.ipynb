###### What is an outlier?

An outlier is a data poiny that is very high or very low compared to other data points. it does not follow the general pattern of the data

```python
data = [10, 12, 11, 13, 200]
# Here 200 is an outlier because most values are around 10–13.
```

```python
Why Handle Outliers?

because outliers can:
= distort the mean and standard deviation
= Affect model accuracy
= Mislead data visualization
= Reduce training performance in machine learning

How to detect Outliers

1) Using IOR (interquartile range)  2) Using Z-score
```

```python
#1)Using IQR (Interquartile Range)
import pandas as pd

data = pd.DataFrame({'Marks': [50, 52, 49, 51, 100]})

Q1 = data['Marks'].quantile(0.25)
Q3 = data['Marks'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

outliers = data[(data['Marks'] < lower_limit) | (data['Marks'] > upper_limit)]
print("Outliers:\n", outliers)

```

###### How to handle Outliers

1) Remove Outliers
Delect extream values (only if they are errors).

```python
filtered_data = data[(data['Marks'] >= lower_limit) & (data['Marks'] <= upper_limit)]
print(filtered_data)
```

###### 2) Capping / flooring (winsorization)

Replace extreame value with the nearest limit.

```python
import numpy as np

data['Marks'] = np.where(data['Marks'] > upper_limit, upper_limit,np.where(data['Marks'] < lower_limit, lower_limit, data['Marks']))
print(data)
```

###### 3) Replace with Mean or Modian 

use when dataset is small or you want to keep the same size

```python
median_value = data['Marks'].median()
data['Marks'] = np.where((data['Marks'] > upper_limit) | (data['Marks'] < lower_limit),median_value, data['Marks'])
print(data)
```

###### 4) Transformation method 

Apply log or square root transformation to reduce extreame effects.

```python
data['Log_Marks'] = np.log(data['Marks'])
print(data)
```

###### 5)Keep Outliers (When Genuine)

Sometimes outliers are real and useful, such as:

High salary
Unusual sales spike
Real-world rare cases
In that case, don’t remove them, just analyze separately.
